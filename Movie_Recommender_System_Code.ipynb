```python
import pandas as pd 
import numpy as np 
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(3)
```

```python
credits.head(3)
```

```python
movies.shape, credits.shape
```

```python
df = movies.merge(credits, on="title")
```

```python
df.shape
```

```python
df.head(2)
```

```python
df.info()
```

```python
df = df[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
import ast
```

```python
def fetch(name):
    L = [] 
    for i in ast.literal_eval(name): 
        L.append(i["name"])
    return L
```

```python
df['genres'] = df['genres'].apply(fetch)
```

```python
df['keywords'] = df['keywords'].apply(fetch)
```

```python
def fetch2(name):
    L = []
    counter = 0
    for i in ast.literal_eval(name): 
        if counter != 5: 
            L.append(i['name'])
            counter += 1
        else: 
            break
    return L
    
```

```python
df['cast'] = df['cast'].apply(fetch2)
```

```python
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
def fetch3(name):
    L = []
    for i in ast.literal_eval(name): 
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
df['crew'] = df['crew'].apply(fetch3)
```

```python
df.head(2)
```

```python
df['cast'] = df['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
df['genres'] = df['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
df['keywords'] = df['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
df['crew'] = df['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
df.head(2)
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
new_df = df[["id", 'title', 'tags']]
```

```python
new_df.head(2)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head(5)
```

```python
new_df['tags'][1]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
    y = []
    for i in obj.split():
         y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out() 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0].shape
```

```python
vector.shape
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
